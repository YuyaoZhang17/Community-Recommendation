```python
areas10=[]
for i in range(22):
    core=cores[i]
    url=part1+"driving/"+core+part2+"10"+part3
    r=requests.get(url)
    a=json.loads(r.text)
    coor=a["features"][0]["geometry"]["coordinates"][0]
    obj={'type':'Polygon','coordinates':[coor]}
    areas10.append(area(obj))
            
```
