```python
import requests 
import pandas as pd
import json
import numpy as np
```

```python
def polygon_area(points):
    area = 0
    q = points[-1]
    for p in points:
        area += p[0] * q[1] - p[1] * q[0]
        q = p
    return area / 2
```

```python
areas10=[]
coors10=[]
for i in range(len(cores)):
    core=cores[i]
    url=part1+"driving/"+core+part2+"10"+part3
    r=requests.get(url)
    a=json.loads(r.text)
    coor=a["features"][0]["geometry"]["coordinates"][0]
    coors10.append(coor)
    polygon=np.array(coor)
    areas10.append(abs(polygon_area(coor)))
```
